```python
from IPython.display import display
```

```python
clinical_cols = [
            'Clinical Annotation ID',
            'Variant/Haplotypes','Gene', 'Drug(s)', 'Phenotype Category', 'Phenotype(s)',
            'Level of Evidence', 'PMID Count', 'Evidence Count'
        ]

drug_cols = [
            'PharmGKB ID',
            'Variants/Haplotypes', 'Genes', 'Chemicals', 'Testing Level', 'Source',
            'Biomarker Flag', 'Has Prescribing Info', 'Has Dosing Info', 'Has Alternate Drug', 'Cancer Genome', 'Prescribing'
        ]
```

```python
def show_drug_gene(drug, gene):
    drug = drug.lower()
    gene = gene.lower()
    display(
        clinical_annotations[(clinical_annotations['Drug(s)'].str.lower().str.contains(drug))
                             & (clinical_annotations['Gene'].str.lower().str.contains(gene))]
        .sort_values('Variant/Haplotypes')[clinical_cols]
    )
    display(
        drug_labels[(drug_labels['Chemicals'].str.lower().str.contains(drug)) 
                    & (drug_labels['Genes'].str.lower().str.contains(gene))]
        .sort_values('Variants/Haplotypes')[drug_cols]
    )
```

```python
show_drug_gene('fluorouracil', 'dpyd')
```

```python
show_drug_gene('fluorouracil', 'tyms')
```
